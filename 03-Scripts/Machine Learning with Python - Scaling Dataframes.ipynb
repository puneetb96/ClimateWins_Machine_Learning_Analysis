## 1. Importing Libraries, Path and Data

```python
#Import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import os
import sklearn
from sklearn.preprocessing import StandardScaler
```

```python
#Import Climate CHange Data Set CSV file
df = pd.read_csv(os.path.join(path, 'Data Sets' , 'Dataset-weather-prediction-dataset-processed.csv'))
```

## 2. Perform data checks

```python
#Check data
df.head()
```

```python
#Create ID column
df.reset_index(inplace=True)
df.rename(columns={'index':'id'}, inplace=True)
df.head()
```

```python
#Check data types
df.dtypes
```

```python
#Create subset to check for duplicates
df_dups = df[df.duplicated()]

#Check duplicate dataset
df_dups.head()
```

#### - No duplicates

```python
#Check for null values
df.isnull().sum()
```

```python
#Remove any nulls
df.dropna(inplace=True)
```

```python
#Check info
df.info
```

## 3. Scale Dataframe

```python
#create scaler object using StandardScaler from sklearn.preprocessing
#StandardScaler assumes data is normally distributed and scales with a distribution
#around 0 and standard deviation of 1. Scaling happens independently with each variable.
scaler = StandardScaler()
```

```python
#Create new dataframe with scaled data
df_scaled = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)

#Check new scaled dataframe
df_scaled.head()
```

```python
#Compare origional data vs scaled
df.head()
```

#### -

## 4. Export Scaled Data

```python
#Export scaled data as PKL file
df_scaled.to_pickle(os.path.join(path, 'Data Sets', 'weather_dataset_scaled.pkl'))
```

```python
#Export sacled data as CSV file
df_scaled.to_csv(os.path.join(path, 'Data Sets', 'weather_dataset_scaled.csv'))
```
